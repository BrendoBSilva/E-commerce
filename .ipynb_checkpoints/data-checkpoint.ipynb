```python
import pandas as pd
import numpy as np

np.random.seed(42)
```

```python
n_customers = 300

customers = pd.DataFrame({
    'customer_id': range(1, n_customers + 1),
    'city': np.random.choice(['São Paulo', 'Rio', 'Belo Horizonte', 'Curitiba'], n_customers),
    'signup_date': pd.to_datetime('2022-01-01') + pd.to_timedelta(np.random.randint(0, 365, n_customers), unit='D')
})

products = pd.DataFrame({
    'product_id': range(1, 21),
    'category': np.random.choice(['Eletrônicos', 'Roupas', 'Casa', 'Esporte'], 20),
    'price': np.random.uniform(50, 500, 20).round(2)
})

n_orders = 2000

orders = pd.DataFrame({
    'order_id': range(1, n_orders + 1),
    'customer_id': np.random.choice(customers['customer_id'], n_orders),
    'product_id': np.random.choice(products['product_id'], n_orders),
    'quantity': np.random.randint(1, 5, n_orders),
    'order_date': pd.to_datetime('2023-01-01') + pd.to_timedelta(np.random.randint(0, 365, n_orders), unit='D'),
    'status': np.random.choice(['pago', 'cancelado'], n_orders, p=[0.9, 0.1])
})

dates = pd.date_range(start='2023-01-01', end='2023-12-31')

ads = pd.DataFrame({
    'date': dates,
    'channel': np.random.choice(['Google', 'Facebook', 'Instagram'], len(dates)),
    'cost': np.random.uniform(100, 1000, len(dates)).round(2),
    'clicks': np.random.randint(100, 1000, len(dates)),
    'conversions': np.random.randint(10, 100, len(dates))
})

orders = orders.merge(products[['product_id', 'price']], on='product_id')

orders['revenue'] = orders['quantity'] * orders['price']

customers.to_csv('customers.csv', index=False)
products.to_csv('products.csv', index=False)
orders.to_csv('orders.csv', index=False)
ads.to_csv('ads.csv', index=False)
```
